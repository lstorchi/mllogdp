```python
# il file MOE_GRID_1078_SCALED_MOE.csv lo ho ottenuto a partire dal file base MOE_GRID_1078.xlsx( vedere inizio foglio descriptors_MOE_SCALED_GRID_NOT_SCALED_scale_false )
import pandas as pd
file = './data/MOE_GRID_1078_SCALED_MOE.csv'
file_read = pd.read_csv(file)
file_dataframe = pd.DataFrame(file_read)
file_dataframe
```

```python
import numpy as np 
import string


list_of_desc = ["a_acc", \
                  "bpol", \
                  "b_double", \
                  "b_rotN", \
                  "b_rotR", \
                  "chi1_C", \
                  "diameter", \
                  "logP(o/w)", \
                  "logS", \
                  "PEOE_VSA+0", \
                  "PEOE_VSA+3", \
                  "PEOE_VSA-0", \
                  "PEOE_VSA-2", \
                  "PEOE_VSA-5", \
                  "PEOE_VSA_POS", \
                  "SlogP_VSA0", \
                  "SlogP_VSA1", \
                  "SlogP_VSA2", \
                  "SlogP_VSA3", \
                  "SlogP_VSA4", \
                  "SlogP_VSA8", \
                  "SMR_VSA1", \
                  "SMR_VSA6", \
                  "TPSA", \
                  "vsa_acc", \
                  "vsa_other", \
                  "vsa_pol", \
                  "weinerPol", \
                  "GRID_DESCRIPTOR"]
                
basicname = ["G", "M", "GM", \
             "GTr", "GTe", \
             "MTe", "MTr", \
             "GMTe", "GMTr"]

X = {}
Y = {}
setid = {}
uniqueid = {}

for bn in basicname:
    X[bn] = []
    Y[bn] = []
    setid[bn] = []
    uniqueid[bn] = []

for idx, ss in enumerate(file_read["UNIQUEID"]):
    
    fdesc = file_read[file_read["UNIQUEID"] == ss]
    
    xval = {}
    for bn in basicname:
        xval[bn] = []

    for d in list_of_desc:
        if d == "GRID_DESCRIPTOR":
            nvs = fdesc[d].values[0].split()
            xval["GM"].extend([float(nv) for nv in nvs])  
            xval["G"].extend([float(nv) for nv in nvs])  
            xval["GMTr"].extend([float(nv) for nv in nvs])  
            xval["GTr"].extend([float(nv) for nv in nvs])  
            xval["GMTe"].extend([float(nv) for nv in nvs])  
            xval["GTe"].extend([float(nv) for nv in nvs])  
        else:
            xval["GM"].append(fdesc[d].values[0])
            xval["M"].append(fdesc[d].values[0])
            xval["GMTr"].append(fdesc[d].values[0])
            xval["MTr"].append(fdesc[d].values[0])
            xval["GMTe"].append(fdesc[d].values[0])
            xval["MTe"].append(fdesc[d].values[0])
            
    y = file_read[file_read["UNIQUEID"] == ss]['LOGD'].values[0]
    sid = file_read[file_read["UNIQUEID"] == ss]['SETID'].values[0]
    uid = file_read[file_read["UNIQUEID"] == ss]['UNIQUEID'].values[0]
    
    for bn in ["GM", "G", "M"]:
        X[bn].append(xval[bn])
        Y[bn].append(y)
        setid[bn].append(sid)
        uniqueid[bn].append(uid)
        
    if (sid == "Tr"):
        for bn in ["GMTr", "GTr", "MTr"]:
            X[bn].append(xval[bn])
            Y[bn].append(y)
            setid[bn].append(sid)
            uniqueid[bn].append(uid)
    elif (sid == "Te"):
        for bn in ["GMTe", "GTe", "MTe"]:
            X[bn].append(xval[bn])
            Y[bn].append(y)
            setid[bn].append(sid)
            uniqueid[bn].append(uid)


for bn in basicname:
    print(bn, " : ")
    print("  ", np.asarray(Y[bn]).shape, np.asarray(X[bn]).shape )  
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=55)
principalComponents = pca.fit_transform(X["G"])
print(pca.explained_variance_ratio_)

# Plot the explained variances
features = range(pca.n_components_)
explained_variance = pca.explained_variance_ratio_
plt.bar(features, explained_variance, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)

plt.show()
```

```python
PCA_components = pd.DataFrame(principalComponents)
PCA_components
```

```python
from sklearn.cluster import KMeans

n_comps_DROP_INIZIALE = 2
n_comps_SECONDO_MINIDROP = 4
ks = range(1, 10)
inertias_2 = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model_2 = KMeans(n_clusters=k)
    # Fit model to samples    
    model_2.fit(PCA_components.iloc[:,:2])
    # Append the inertia to the list of inertias
    inertias_2.append(model_2.inertia_)
    
plt.plot(ks, inertias_2, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
import numpy as np

# Using three clusters 
NC = 3

model = KMeans(n_clusters=NC)
   
X_2 = PCA_components.iloc[:,:2].values

y_km = model.fit(X_2)

plt.scatter(PCA_components[0], PCA_components[1], alpha=.1, color='black')

clusters = {}
for idx , l in enumerate(y_km.labels_):
    if l in clusters:
        clusters[l].append(X_2[idx])
    else:
        clusters[l] = []
        clusters[l].append(X_2[idx])
        
        
c1 = np.asarray(clusters[0])
plt.scatter(
    c1[:,0],c1[:,1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

c2 = np.asarray(clusters[1])
plt.scatter(
    c2[:,0],c2[:,1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

c3 = np.asarray(clusters[2])
plt.scatter(
    c3[:,0],c3[:,1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

# plot the centroids
plt.scatter(
    model.cluster_centers_[:, 0], model.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)

plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import cross_val_predict
import matplotlib.pyplot as plt
import numpy as np
import matplotlib 
import math

def componentiPLS (X_Tr, Y_Tr, cvval, maxcomp):
    
    print("Running GRID model using cv = %3d"%(cvval))
    
    trscores_cv = []
    trncomps =[]
    trmses_cv = []
    
    for trncomp in range(1,maxcomp+1):
            pls = PLSRegression(trncomp, scale=False)
            pls.fit( X_Tr, Y_Tr )
        
            y_cv = cross_val_predict(pls, X_Tr, Y_Tr, cv=cvval)
            trscore_cv = r2_score(Y_Tr, y_cv)
            trmse_cv = mean_squared_error(Y_Tr, y_cv)
        
            trscores_cv.append(trscore_cv)
            trmses_cv.append(trmse_cv)
            trncomps.append(trncomp)
            print("%4d %10.8f %10.8f"%(trncomp, trmse_cv, trscore_cv))
    
    msetrmin = np.argmin(trmses_cv)
    
    plt.plot(trncomps, trscores_cv, '-o', color='black')
    plt.xlabel('number of components tr')
    plt.ylabel('scores tr')
    plt.xticks(trncomps)
    plt.show()
    
    plt.plot(trncomps, trmses_cv, '-o', color='black')
    plt.xlabel('number of components tr')
    plt.ylabel('MSE tr')
    plt.xticks(trncomps)
    plt.show()

    acomp = msetrmin+1
    print("Number of components selected automatically: ",  acomp)
```

```python
def ModelliAutoManuali( X_Tr, Y_Tr, X_Te, Y_Te, cvval, auto, manual) :
    pls_auto = PLSRegression(auto, scale=False)
    pls_auto.fit(X_Tr, Y_Tr)
    y_auto_te = pls_auto.predict(X_Te)
    y_auto_tr = pls_auto.predict(X_Tr)
    y_auto = np.concatenate((y_auto_te, y_auto_tr))
    y_auto_true = []
    y_auto_true.extend(Y_Te)
    y_auto_true.extend(Y_Tr)
        
    diff_auto = []
    diffnoa_auto = []
    for idx, v in enumerate(y_auto_true):
        d_auto = math.fabs(v-y_auto[idx])
        diff_auto.append(d_auto)
        d_noa_auto = v-y_auto[idx]
        diffnoa_auto.append(d_noa_auto)
            
    print( np.average(diff_auto), np.std(diff_auto), np.std(diffnoa_auto) )
        
    pls_manu = PLSRegression(manual, scale=False)
    pls_manu.fit(X_Tr, Y_Tr)
    y_manu_te = pls_manu.predict(X_Te)
    y_manu_tr = pls_manu.predict(X_Tr)
    y_manu = np.concatenate((y_manu_te, y_manu_tr))
    y_manu_true = []
    y_manu_true.extend(Y_Te)
    y_manu_true.extend(Y_Tr)
        
    diff_manual = []
    diffnoa_manual = []
    for idx, v in enumerate(y_manu_true):
        d_manu = math.fabs(v-y_manu[idx])
        diff_manual.append(d_manu)
        d_noa_manu = v-y_manu[idx]
        diffnoa_manual.append(d_noa_manu)
            
    print( np.average(diff_manual), np.std(diff_manual), np.std(diffnoa_manual) )
    
    plt.scatter(y_auto_true, y_auto)
    plt.xlabel("Y_auto_true")
    plt.ylabel("Y_auto")
    plt.show()
        
    plt.scatter(y_manu_true, y_manu)
    plt.xlabel("Y_manu_true")
    plt.ylabel("Y_manu")
    plt.show()
    
    plt.scatter(Y_Tr, y_auto_tr)
    plt.xlabel("Y_vera_Tr")
    plt.ylabel("Y_predetta_auto_Tr")
    plt.show()
    
    plt.scatter(Y_Tr, y_manu_tr)
    plt.xlabel("Y_vera_Tr")
    plt.ylabel("Y_predetta_manu_Tr")
    plt.show()
    
    plt.scatter(Y_Te, y_auto_te)
    plt.xlabel("Y_vera_Te")
    plt.ylabel("Y_predetta_auto_Te")
    plt.show()
    
    plt.scatter(Y_Te, y_manu_te)
    plt.xlabel("Y_vera_Te")
    plt.ylabel("Y_predetta_manu_Te")
    plt.show()
```
