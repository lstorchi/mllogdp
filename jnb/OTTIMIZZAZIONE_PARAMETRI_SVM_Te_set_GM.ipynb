```python
%%time
import pandas as pd
file = './data/MOE_GRID_1078.xlsx'
file_read = pd.read_excel(file)
file_dataframe = pd.DataFrame(file_read)
file_dataframe
```

```python
%%time
import numpy as np 
import string


list_of_desc = ["a_acc", \
                  "bpol", \
                  "b_double", \
                  "b_rotN", \
                  "b_rotR", \
                  "chi1_C", \
                  "diameter", \
                  "logP(o/w)", \
                  "logS", \
                  "PEOE_VSA+0", \
                  "PEOE_VSA+3", \
                  "PEOE_VSA-0", \
                  "PEOE_VSA-2", \
                  "PEOE_VSA-5", \
                  "PEOE_VSA_POS", \
                  "SlogP_VSA0", \
                  "SlogP_VSA1", \
                  "SlogP_VSA2", \
                  "SlogP_VSA3", \
                  "SlogP_VSA4", \
                  "SlogP_VSA8", \
                  "SMR_VSA1", \
                  "SMR_VSA6", \
                  "TPSA", \
                  "vsa_acc", \
                  "vsa_other", \
                  "vsa_pol", \
                  "weinerPol", \
                  "DESCRIPTOR"]
                
basicname = ["G", "M", "GM", \
             "GTr", "GTe", \
             "MTe", "MTr", \
             "GMTe", "GMTr"]

X = {}
Y = {}
setid = {}
uniqueid = {}

for bn in basicname:
    X[bn] = []
    Y[bn] = []
    setid[bn] = []
    uniqueid[bn] = []

for idx, ss in enumerate(file_read["UNIQUEID"]):
    
    fdesc = file_read[file_read["UNIQUEID"] == ss]
    
    xval = {}
    for bn in basicname:
        xval[bn] = []

    for d in list_of_desc:
        if d == "DESCRIPTOR":
            nvs = fdesc[d].values[0].split()
            xval["GM"].extend([float(nv) for nv in nvs])  
            xval["G"].extend([float(nv) for nv in nvs])  
            xval["GMTr"].extend([float(nv) for nv in nvs])  
            xval["GTr"].extend([float(nv) for nv in nvs])  
            xval["GMTe"].extend([float(nv) for nv in nvs])  
            xval["GTe"].extend([float(nv) for nv in nvs])  
        else:
            xval["GM"].append(fdesc[d].values[0])
            xval["M"].append(fdesc[d].values[0])
            xval["GMTr"].append(fdesc[d].values[0])
            xval["MTr"].append(fdesc[d].values[0])
            xval["GMTe"].append(fdesc[d].values[0])
            xval["MTe"].append(fdesc[d].values[0])
            
    y = file_read[file_read["UNIQUEID"] == ss]['LOGD'].values[0]
    sid = file_read[file_read["UNIQUEID"] == ss]['SETID'].values[0]
    uid = file_read[file_read["UNIQUEID"] == ss]['UNIQUEID'].values[0]
    
    for bn in ["GM", "G", "M"]:
        X[bn].append(xval[bn])
        Y[bn].append(y)
        setid[bn].append(sid)
        uniqueid[bn].append(uid)
        
    if (sid == "Tr"):
        for bn in ["GMTr", "GTr", "MTr"]:
            X[bn].append(xval[bn])
            Y[bn].append(y)
            setid[bn].append(sid)
            uniqueid[bn].append(uid)
    elif (sid == "Te"):
        for bn in ["GMTe", "GTe", "MTe"]:
            X[bn].append(xval[bn])
            Y[bn].append(y)
            setid[bn].append(sid)
            uniqueid[bn].append(uid)


for bn in basicname:
    print(bn, " : ")
    print("  ", np.asarray(Y[bn]).shape, np.asarray(X[bn]).shape )  
```

```python
%%time
import sklearn 
import sklearn.preprocessing 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_GM_SCALED_Tr = scaler.fit_transform(X['GMTr'])
X_GM_SCALED_Tr_dataframe = pd.DataFrame(X_GM_SCALED_Tr)
X_GM_SCALED_Tr_dataframe

X_GM_SCALED_Te = scaler.fit_transform(X['GMTe'])
X_GM_SCALED_Te_dataframe = pd.DataFrame(X_GM_SCALED_Te)
X_GM_SCALED_Tr_dataframe
X_GM_SCALED_Te_dataframe
```

```python
%%time
Y_GM_Tr = Y['GMTr']
Y_GM_Te = Y['GMTe']
```

```python
%%time
from sklearn import svm
from sklearn.svm import SVR
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_predict

gamma = ['scale','auto']
kernel = ['linear','sigmoid','rbf']
degree = [2,3,4,5,6,7,8,9,10]
C = [1,50,64,100,1000]
epsilon = [0.1,0.25,0.50,0.75,1.0]
poly_val = ['poly'] 

dizionario_lin_sig_rbf = {}
for g_val in gamma :
    for k_val in kernel:
            for C_val in C :
                for e_val in epsilon :
                    svr_gamma = svm.SVR(gamma=g_val,kernel=k_val, C=C_val, epsilon=e_val)
                    svr_gamma.fit(X_GM_SCALED_Tr,Y_GM_Tr)
                    Y_pred_Te = svr_gamma.predict(X_GM_SCALED_Te)
                    mse = mean_squared_error(Y_GM_Te, Y_pred_Te)
                    r2 = r2_score(Y_GM_Te,Y_pred_Te)
                    dizionario_lin_sig_rbf[g_val+"_"+k_val+"_"+str(e_val)+"_"+str(C_val)] = [mse, r2]    
```

```python
%%time
print(dizionario_lin_sig_rbf)
```

```python
%%time
dizionario_poly = {}
for g_val in gamma :
    for p_val in poly_val:
        for d_val in degree:
            for C_val in C :
                for e_val in epsilon :
                    svr_poly = svm.SVR(gamma=g_val,kernel=p_val,degree=d_val, C=C_val, epsilon=e_val)
                    svr_poly.fit(X_GM_SCALED_Tr,Y_GM_Tr)
                    Y_pred_Te_poly = svr_poly.predict(X_GM_SCALED_Te)
                    mse_poly = mean_squared_error(Y_GM_Te, Y_pred_Te_poly)
                    r2_poly = r2_score(Y_GM_Te,Y_pred_Te_poly)
                    dizionario_poly[g_val+"_"+p_val+"_"+str(d_val)+"_"+str(e_val)+"_"+str(C_val)] = [mse_poly, r2_poly]    
```

```python
%%time
print(dizionario_poly)
```

```python
SPERIAMO = list(dizionario_lin_sig_rbf.keys())
mse_r2 = list(dizionario_lin_sig_rbf.values())
```

```python
mse_vals = []
r2_vals = []
for idx,v in enumerate (mse_r2):
    mse_vals.append(v[0])
    r2_vals.append(v[1])
    
print("mse:\n", mse_vals)
print("r2_vals:\n", r2_vals)
```

```python
DATAFRAME_SPLITTATO_GM = pd.DataFrame(SPERIAMO, columns=['hyperparameters'])
DATAFRAME_SPLITTATO_GM['mse'] = mse_vals
DATAFRAME_SPLITTATO_GM['r2'] = r2_vals
DATAFRAME_SPLITTATO_GM
```
