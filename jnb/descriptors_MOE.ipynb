```python
X = []
Y = []
setid = []
UNIQUEID = []
for idx, ss in enumerate(merged_dataframe_MOEnostri["UNIQUEID"]):
    
    fdesc = merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == ss]
    
    xval = []
    for d in list_of_desc:
        xval.append(fdesc[d].values[0])

    X.append(xval)
    Y.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == ss][' LOGD'].values[0])
    setid.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == ss][' SETID'].values[0])
    UNIQUEID.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == ss]['UNIQUEID'].values[0])
    print(len(Y))
    
for idx, y in enumerate(Y):
    print(setid[idx], X[idx], y)
```

```python
Xtr = []
Xte = []
Ytr = []
Yte = []
for idx, y in enumerate(Y):
    if setid[idx] == " Tr ":
        Xtr.append(X[idx])
        Ytr.append(Y[idx])
    elif setid[idx] == " Te ":
        Xte.append(X[idx])
        Yte.append(Y[idx])
        
print("    Test set: ", len(Yte), len(Xte))
print("Training set: ", len(Ytr), len(Xtr))
```

```python
# valutazione del numero di componenti principali ideale da usare per costruire il modello PLS (SENZA il descrittore nmol )
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_predict
import numpy as np
import matplotlib.pyplot as plt

trscores = []
trmses = []
trncomps = []
for trncomp in range(1,20):
    pls = PLSRegression(trncomp)
    pls.fit(Xtr, Ytr)
    
    y_cv = cross_val_predict(pls, Xtr, Ytr, cv=10)
    
    trscore = r2_score(Ytr, y_cv)
    trmse = mean_squared_error(Ytr, y_cv)
    
    trscores.append(trscore)
    trmses.append(trmse)
    trncomps.append(trncomp)
    print("%4d %10.8f %10.8f"%(trncomp, trmse, trscore))
    
msetrmin = np.argmin(trmses)
print("Suggested number of components tr: ", msetrmin+1)
    
plt.plot(trncomps, trscores, '-o', color='black')
plt.xlabel('number of components tr')
plt.ylabel('scores tr')
plt.xticks(trncomps)
plt.show()

plt.plot(trncomps, trmses, '-o', color='black')
plt.xlabel('number of components tr')
plt.ylabel('MSE tr')
plt.xticks(trncomps)
plt.show()
```

```python
# NUMERO DI COMPONENTI SUGGERITO : 15
# NUMERO DI COMPONENTI DOPO VALUTAZIONE OCCHIOMETRICA : 8
```

```python
ncompsugg = 15

pls_15 = PLSRegression(ncompsugg)
pls_15.fit(Xtr, Ytr)
    
y_cv15 = cross_val_predict(pls_15, Xtr, Ytr, cv=10)
    
score15cv = r2_score(Ytr, y_cv15)
mse15cv = mean_squared_error(Ytr, y_cv15)    

print("comp: %4d MSE: %10.8f score: %10.8f"%(ncompsugg, mse15cv, score15cv))
```

```python
ncompocchio = 8

pls_8 = PLSRegression(ncompocchio)
pls_8.fit(Xtr, Ytr)
    
y_cv8 = cross_val_predict(pls_8, Xtr, Ytr, cv=10)
    
score8cv = r2_score(Ytr, y_cv8)
mse8cv = mean_squared_error(Ytr, y_cv8)    

print("comp: %4d MSE: %10.8f score: %10.8f"%(ncompocchio, mse8cv, score8cv))
```

```python
Y_pred15 = pls_15.predict(Xte)
Y_pred8 = pls_8.predict(Xte)

Y_pred15_TR = pls_15.predict(Xtr)
Y_pred8_TR = pls_8.predict(Xtr)


mse_pred15 = mean_squared_error(Yte, Y_pred15 )

RMSE_pred15 = mean_squared_error(Yte, Y_pred15, squared = False )

mse_pred8 = mean_squared_error(Yte, Y_pred8 )

RMSE_pred8 = mean_squared_error(Yte, Y_pred8, squared = False )


diff15 = []
diff8 =[]

for idx, val in enumerate(Yte):
    diff15.append(val-Y_pred15[idx])
    diff8.append(val-Y_pred8[idx])
    print("%10.7f %10.7f %10.7f %10.7f %10.7f"%(\
                                val, Y_pred15[idx], diff15[-1] ,Y_pred8[idx], diff8[-1] )) 
                            
score15 = r2_score(Yte, Y_pred15 )

score8 = r2_score(Yte, Y_pred8 )


print(" score modello a 15 componenti :", score15 )
print(" mse modello a 15 componenti :", mse_pred15)

print(" RMSE modello a 15 componenti :", RMSE_pred15)

print(" score modello a 8 componenti :", score8 )
print(" mse modello a 8 componenti :", mse_pred8)

print(" RMSE modello a 8 componenti :", RMSE_pred8)



plt.xlabel(" True Y value")
plt.ylabel(" Predicted Y values model built using 15 components) ")

plt.scatter(Yte, Y_pred15)

plt.show()

plt.xlabel(" True Y value")
plt.ylabel(" Predicted Y values model built using 8 components) ")

plt.scatter(Yte, Y_pred8)

plt.show()
```
