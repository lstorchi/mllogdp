```python
%%time
import pandas as pd
file = './data/MOE_GRID_1078.xlsx'
file_read = pd.read_excel(file)
file_dataframe = pd.DataFrame(file_read)
file_dataframe
```

```python
%%time
import numpy as np 
import string


list_of_desc = ["a_acc", \
                  "bpol", \
                  "b_double", \
                  "b_rotN", \
                  "b_rotR", \
                  "chi1_C", \
                  "diameter", \
                  "logP(o/w)", \
                  "logS", \
                  "PEOE_VSA+0", \
                  "PEOE_VSA+3", \
                  "PEOE_VSA-0", \
                  "PEOE_VSA-2", \
                  "PEOE_VSA-5", \
                  "PEOE_VSA_POS", \
                  "SlogP_VSA0", \
                  "SlogP_VSA1", \
                  "SlogP_VSA2", \
                  "SlogP_VSA3", \
                  "SlogP_VSA4", \
                  "SlogP_VSA8", \
                  "SMR_VSA1", \
                  "SMR_VSA6", \
                  "TPSA", \
                  "vsa_acc", \
                  "vsa_other", \
                  "vsa_pol", \
                  "weinerPol", \
                  "DESCRIPTOR"]
                
basicname = ["G", "M", "GM", \
             "GTr", "GTe", \
             "MTe", "MTr", \
             "GMTe", "GMTr"]

X = {}
Y = {}
setid = {}
uniqueid = {}

for bn in basicname:
    X[bn] = []
    Y[bn] = []
    setid[bn] = []
    uniqueid[bn] = []

for idx, ss in enumerate(file_read["UNIQUEID"]):
    
    fdesc = file_read[file_read["UNIQUEID"] == ss]
    
    xval = {}
    for bn in basicname:
        xval[bn] = []

    for d in list_of_desc:
        if d == "DESCRIPTOR":
            nvs = fdesc[d].values[0].split()
            xval["GM"].extend([float(nv) for nv in nvs])  
            xval["G"].extend([float(nv) for nv in nvs])  
            xval["GMTr"].extend([float(nv) for nv in nvs])  
            xval["GTr"].extend([float(nv) for nv in nvs])  
            xval["GMTe"].extend([float(nv) for nv in nvs])  
            xval["GTe"].extend([float(nv) for nv in nvs])  
        else:
            xval["GM"].append(fdesc[d].values[0])
            xval["M"].append(fdesc[d].values[0])
            xval["GMTr"].append(fdesc[d].values[0])
            xval["MTr"].append(fdesc[d].values[0])
            xval["GMTe"].append(fdesc[d].values[0])
            xval["MTe"].append(fdesc[d].values[0])
            
    y = file_read[file_read["UNIQUEID"] == ss]['LOGD'].values[0]
    sid = file_read[file_read["UNIQUEID"] == ss]['SETID'].values[0]
    uid = file_read[file_read["UNIQUEID"] == ss]['UNIQUEID'].values[0]
    
    for bn in ["GM", "G", "M"]:
        X[bn].append(xval[bn])
        Y[bn].append(y)
        setid[bn].append(sid)
        uniqueid[bn].append(uid)
        
    if (sid == "Tr"):
        for bn in ["GMTr", "GTr", "MTr"]:
            X[bn].append(xval[bn])
            Y[bn].append(y)
            setid[bn].append(sid)
            uniqueid[bn].append(uid)
    elif (sid == "Te"):
        for bn in ["GMTe", "GTe", "MTe"]:
            X[bn].append(xval[bn])
            Y[bn].append(y)
            setid[bn].append(sid)
            uniqueid[bn].append(uid)


for bn in basicname:
    print(bn, " : ")
    print("  ", np.asarray(Y[bn]).shape, np.asarray(X[bn]).shape )  
```

```python
%%time
import sklearn 
import sklearn.preprocessing 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_M_SCALED_Tr = scaler.fit_transform(X['MTr'])
X_M_SCALED_Tr_dataframe = pd.DataFrame(X_M_SCALED_Tr)
X_M_SCALED_Tr_dataframe

X_M_SCALED_Te = scaler.fit_transform(X['MTe'])
X_M_SCALED_Te_dataframe = pd.DataFrame(X_M_SCALED_Te)
print(X_M_SCALED_Tr_dataframe)
print(X_M_SCALED_Te_dataframe)

X_M_SCALED_TOT = scaler.fit_transform(X['M'])
X_M_SCALED_TOT_dataframe = pd.DataFrame(X_M_SCALED_TOT)
print(X_M_SCALED_TOT_dataframe)
```

```python
%%time
Y_M_Tr = Y['MTr']
Y_M_Te = Y['MTe']
Y_M_TOT = Y['M']
```

```python
%%time
import math
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.svm import SVR
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_predict 

def r2_RMSE_diff_abs_label( Y_true, Y_pred, label):
    r2_val = r2_score( Y_true, Y_pred )
    print( " Il valore di r2 sul",label,"set è:", r2_val)
    RMSE_val = mean_squared_error( Y_true, Y_pred, squared=False )
    print( " Il valore di RMSE sul",label,"set è:", RMSE_val)
    
    for idx, v in enumerate(Y_true):
        diff_abs = math.fabs(v-Y_pred[idx])
        diff_abs_modello.append(diff_abs)
        diff_NOabs = v-Y_pred[idx]
        diff_NOabs_modello.append(diff_NOabs)
    
    print( "Gli errori sul", label," set sono :",np.average(diff_abs_modello), np.std(diff_abs_modello), np.std(diff_NOabs_modello) )
    


svr_model = svm.SVR(gamma='scale',kernel='rbf', C=64, epsilon=0.5)
svr_model.fit(X_M_SCALED_Tr,Y_M_Tr)
Y_pred_TOT = svr_model.predict(X_M_SCALED_TOT)
Y_pred_Tr = svr_model.predict(X_M_SCALED_Tr)
Y_pred_Te = svr_model.predict(X_M_SCALED_Te)
diff_abs_modello = []
diff_NOabs_modello = []
    
for idx, v in enumerate(Y_M_TOT):
    diff_abs = math.fabs(v-Y_pred_TOT[idx])
    diff_abs_modello.append(diff_abs)
    diff_NOabs = v-Y_pred_TOT[idx]
    diff_NOabs_modello.append(diff_NOabs)
    
print( np.average(diff_abs_modello), np.std(diff_abs_modello), np.std(diff_NOabs_modello) )              
    
plt.scatter(Y_M_TOT, Y_pred_TOT)
plt.xlabel("Y_true_TOT")
plt.ylabel("Y_pred_TOT")
plt.show()   
    
plt.scatter(Y_M_Tr, Y_pred_Tr)
plt.xlabel("Y_true_Tr")
plt.ylabel("Y_pred_Tr")
plt.show() 
    
plt.scatter(Y_M_Te, Y_pred_Te)
plt.xlabel("Y_true_Te")
plt.ylabel("Y_pred_Te")
plt.show() 
r2_Te = r2_score(Y_M_Te, Y_pred_Te)
print("r2_Te =", r2_Te)
```
