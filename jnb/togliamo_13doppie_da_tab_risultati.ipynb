```python
import pandas as pd
import numpy as np
filename ='./data/confronto.csv'
confronto = pd.read_csv(filename)
confronto_dataframe = pd.DataFrame(confronto)

```

```python
filename = './data/1130_alldesc_notdormol.csv'
all_desc = pd.read_csv(filename)
all_desc_dataframe = pd.DataFrame(all_desc)
all_desc_dataframe
```

```python
new_tab = pd.DataFrame(all_desc_dataframe[['mol','Unique_ID', 'nmol']])
new_tab
```

```python
merged_dataframe = pd.merge(confronto_dataframe, new_tab, left_on='NO', right_on='Unique_ID')
merged_dataframe
print(merged_dataframe.shape)
```

```python
merged_dataframe = merged_dataframe[merged_dataframe["nmol"] != 2]
merged_dataframe
print(merged_dataframe.shape)
```
