```r
install.packages("readxl")
library( readxl )
```

```r
file <- '/Users/gdt31/Downloads/GRAFICI ANDAMENTO DESCRITTORI MOE.xlsx'
file_table <- read_excel( file )
file_table
```

```r
# numero di colonne nella tabella
conta <- 0
for ( c in file_table){
    conta <- conta+1
}
print( conta )    
```

```r
nomi_colonne <- c( "a_acc", "a_base","bpol","b_double","b_rotN","b_rotR","chi1_C","diameter","logP(o/w)","logS","nmol","PEOE_VSA+0","PEOE_VSA+3","PEOE_VSA-0","PEOE_VSA-2","PEOE_VSA-5","PEOE_VSA_POS","SlogP_VSA0","SlogP_VSA1","SlogP_VSA2","SlogP_VSA3","SlogP_VSA4","SlogP_VSA8","SMR_VSA1","SMR_VSA6","TPSA","vsa_acc","vsa_other","vsa_pol","weinerPol")
for ( n in nomi_colonne ){
    print( summary( file_table[n]) )
}
```

```r
# deviazione standard a_acc
desc <- file_table["a_acc"]
desc_vect <- c()
for ( v in desc ){
    desc_vect <- c(desc_vect,v)
}
cat("dev.st. a_acc :", sd( desc_vect ) )
hist( desc_vect, col=c("blue") )
plot( desc_vect, type = 'l', col=c("blue"))
```
