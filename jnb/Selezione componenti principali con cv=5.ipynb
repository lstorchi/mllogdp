```python
import pandas as pd 

filename = "./data/1130_alldesc.csv"

descdata = pd.read_csv(filename)

filename = "./data/data.xlsx"

ydata = pd.read_excel(filename)

if len(ydata) != len(descdata):
    print("Error")

print(ydata.head())
```

```python

list_of_desc = ["a_acc", \
                  "a_base", \
                  "bpol", \
                  "b_double", \
                  "b_rotN", \
                  "b_rotR", \
                  "chi1_C", \
                  "diameter", \
                  "logP(o/w)", \
                  "logS", \
                  "nmol", \
                  "PEOE_VSA+0", \
                  "PEOE_VSA+3", \
                  "PEOE_VSA-0", \
                  "PEOE_VSA-2", \
                  "PEOE_VSA-5", \
                  "PEOE_VSA_POS", \
                  "SlogP_VSA0", \
                  "SlogP_VSA1", \
                  "SlogP_VSA2", \
                  "SlogP_VSA3", \
                  "SlogP_VSA4", \
                  "SlogP_VSA8", \
                  "SMR_VSA1", \
                  "SMR_VSA6", \
                  "TPSA", \
                  "vsa_acc", \
                  "vsa_other", \
                  "vsa_pol", \
                  "weinerPol"]

X = []
Y = []
setid = []
for idx, ss in enumerate(ydata["NO"]):
    
    fdesc = descdata[descdata["Unique_ID"] == ss+".mol"]
    
    xval = []
    for d in list_of_desc:
        xval.append(fdesc[d].values[0])

    X.append(xval)
    Y.append(ydata[ydata["NO"] == ss]["LogD"].values[0])
    setid.append(ydata[ydata["NO"] == ss]["DatasetID"].values[0])
    
for idx, y in enumerate(Y):
    print(setid[idx], X[idx], y)
```

```python
Xtr = []
Xte = []
Ytr = []
Yte = []
for idx, y in enumerate(Y):
    if setid[idx] == "Tr":
        Xtr.append(X[idx])
        Ytr.append(Y[idx])
    elif setid[idx] == "Te":
        Xte.append(X[idx])
        Yte.append(Y[idx])
        
print("    Test set: ", len(Yte), len(Xte))
print("Training set: ", len(Ytr), len(Xtr))
```

```python
# valutazione del numero di componenti principali ideale da usare per costruire il modello PLS con cv=5
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_predict
import numpy as np
import matplotlib.pyplot as plt

trscores = []
trmses = []
trncomps = []
for trncomp in range(1,30):
    pls = PLSRegression(trncomp)
    pls.fit(Xtr, Ytr)
    
    y_cv5 = cross_val_predict(pls, Xtr, Ytr, cv=5)
    
    trscore = r2_score(Ytr, y_cv5)
    trmse = mean_squared_error(Ytr, y_cv5)
    
    trscores.append(trscore)
    trmses.append(trmse)
    trncomps.append(trncomp)
    print("%4d %10.8f %10.8f"%(trncomp, trmse, trscore))
    
msetrmin = np.argmin(trmses)
print("Suggested number of components tr: ", msetrmin+1)
    
plt.plot(trncomps, trscores, '-o', color='black')
plt.xlabel('number of components tr')
plt.ylabel('scores tr')
plt.xticks(trncomps)
plt.show()

plt.plot(trncomps, trmses, '-o', color='black')
plt.xlabel('number of components tr')
plt.ylabel('MSE tr')
plt.xticks(trncomps)
plt.show()
```

```python
# costruzione del modello a 20 comps 
ncompsugg = 20

pls_20 = PLSRegression(ncompsugg)
pls_20.fit(Xtr, Ytr)
    
y_cv20 = cross_val_predict(pls_20, Xtr, Ytr, cv=5)
    
score20cv = r2_score(Ytr, y_cv20)
mse20cv = mean_squared_error(Ytr, y_cv20)    

print("comp: %4d MSE: %10.8f score: %10.8f"%(ncompsugg, mse20cv, score20cv))
```

```python
# costruzione del modello a 10 comps
ncompgraf = 10

pls_10 = PLSRegression(ncompgraf)
pls_10.fit(Xtr, Ytr)
    
y_cv10 = cross_val_predict(pls_10, Xtr, Ytr, cv=5)
    
score10cv = r2_score(Ytr, y_cv10)
mse10cv = mean_squared_error(Ytr, y_cv10)    

print("comp: %4d MSE: %10.8f score: %10.8f"%(ncompgraf, mse10cv, score10cv))
```

```python
# predizione dei valori di logD usando modelli costruiti rispettivamente con 20 e 10 comps, calcolo della differenza tra i valori veri di logD e i valori predetti dai 2 modelli 
Y_pred20 = pls_20.predict(Xte)
Y_pred10 = pls_10.predict(Xte)

mse_pred20 = mean_squared_error(Yte, Y_pred20 )
mse_pred10 = mean_squared_error(Yte, Y_pred10 )
RMSE_pred20 = mean_squared_error(Yte, Y_pred20, squared = False )
RMSE_pred10 = mean_squared_error(Yte, Y_pred10, squared = False )

diff10 = []
diff20 = []

for idx, val in enumerate(Yte):
    diff10.append(val-Y_pred10[idx])
    diff20.append(val-Y_pred20[idx])
    print("%10.7f %10.7f %10.7f %10.7f %10.7f"%(\
                                val, Y_pred20[idx], diff20[-1], \
                            Y_pred10[idx], diff10[-1]))
    
print("MSE 20 comp: ", mse_pred20) 
print("MSE 10 comp: ", mse_pred10)
score20 = r2_score(Yte, Y_pred20 )
print(" score modello a 20 componenti :", score20 )
score10 = r2_score(Yte, Y_pred10 )
print(" score modello a 10 componenti :", score10 )
print(" RMSE modello a 20 componenti :", RMSE_pred20)
print(" RMSE modello a 10 componenti :", RMSE_pred10)


plt.xlabel("True Y value")
plt.ylabel("Predicted Y values ( model built using 20 components)")

plt.scatter(Yte, Y_pred20)

plt.show()

plt.xlabel(" True Y value")
plt.ylabel(" Predicted Y values (model built using 10 components) ")

plt.scatter(Yte, Y_pred10)

plt.show()
```

```python
# GRAFICO : RESIDUI ORDINARI modello a 20 vs Yte
plt.xlabel("Yte")
plt.ylabel("diff20")
plt.scatter( Yte, diff20)
plt.show()
# GRAFICO : RESIDUI ORDINARI modello a 10 vs Yte
plt.xlabel("Yte")
plt.ylabel("diff10")
plt.scatter( Yte, diff10)
plt.show()
```
