```python
import pandas as pd 

filename = "./data/1130_alldesc.csv"

descdata = pd.read_csv(filename)

filename = "./data/data.xlsx"

ydata = pd.read_excel(filename)

if len(ydata) != len(descdata):
    print("Error")

print(ydata.head())
```

```python
list_of_desc = ["a_acc", \
                  "a_base", \
                  "bpol", \
                  "b_double", \
                  "b_rotN", \
                  "b_rotR", \
                  "chi1_C", \
                  "diameter", \
                  "logP(o/w)", \
                  "logS", \
                  "PEOE_VSA+0", \
                  "PEOE_VSA+3", \
                  "PEOE_VSA-0", \
                  "PEOE_VSA-2", \
                  "PEOE_VSA-5", \
                  "PEOE_VSA_POS", \
                  "SlogP_VSA0", \
                  "SlogP_VSA1", \
                  "SlogP_VSA2", \
                  "SlogP_VSA3", \
                  "SlogP_VSA4", \
                  "SlogP_VSA8", \
                  "SMR_VSA1", \
                  "SMR_VSA6", \
                  "TPSA", \
                  "vsa_acc", \
                  "vsa_other", \
                  "vsa_pol", \
                  "weinerPol"]

X = []
Y = []
setid = []
for idx, ss in enumerate(ydata["NO"]):
    
    fdesc = descdata[descdata["Unique_ID"] == ss+".mol"]
    
    xval = []
    for d in list_of_desc:
        xval.append(fdesc[d].values[0])

    X.append(xval)
    Y.append(ydata[ydata["NO"] == ss]["LogD"].values[0])
    setid.append(ydata[ydata["NO"] == ss]["DatasetID"].values[0])
    
for idx, y in enumerate(Y):
    print(setid[idx], X[idx], y)
```

```python
Xtr = []
Xte = []
Ytr = []
Yte = []
for idx, y in enumerate(Y):
    if setid[idx] == "Tr":
        Xtr.append(X[idx])
        Ytr.append(Y[idx])
    elif setid[idx] == "Te":
        Xte.append(X[idx])
        Yte.append(Y[idx])
        
print("    Test set: ", len(Yte), len(Xte))
print("Training set: ", len(Ytr), len(Xtr))
```

```python
# valutazione del numero di componenti principali ideale da usare per costruire il modello PLS (SENZA il descrittore nmol )
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_predict
import numpy as np
import matplotlib.pyplot as plt

trscores = []
trmses = []
trncomps = []
for trncomp in range(1,30):
    pls = PLSRegression(trncomp)
    pls.fit(Xtr, Ytr)
    
    y_cv = cross_val_predict(pls, Xtr, Ytr, cv=10)
    
    trscore = r2_score(Ytr, y_cv)
    trmse = mean_squared_error(Ytr, y_cv)
    
    trscores.append(trscore)
    trmses.append(trmse)
    trncomps.append(trncomp)
    print("%4d %10.8f %10.8f"%(trncomp, trmse, trscore))
    
msetrmin = np.argmin(trmses)
print("Suggested number of components tr: ", msetrmin+1)
    
plt.plot(trncomps, trscores, '-o', color='black')
plt.xlabel('number of components tr')
plt.ylabel('scores tr')
plt.xticks(trncomps)
plt.show()

plt.plot(trncomps, trmses, '-o', color='black')
plt.xlabel('number of components tr')
plt.ylabel('MSE tr')
plt.xticks(trncomps)
plt.show()
```

```python
ncompsugg_senzanmol = 21

pls_21senzanmol = PLSRegression(ncompsugg_senzanmol)
pls_21senzanmol.fit(Xtr, Ytr)
    
y_cv21senzanmol = cross_val_predict(pls_21senzanmol, Xtr, Ytr, cv=10)
    
score21cv = r2_score(Ytr, y_cv21senzanmol)
mse21cv = mean_squared_error(Ytr, y_cv21senzanmol)    

print("comp: %4d MSE: %10.8f score: %10.8f"%(ncompsugg_senzanmol, mse21cv, score21cv))
```

```python
ncompgraf_senzanmol = 12

pls_12senzanmol = PLSRegression(ncompgraf_senzanmol)
pls_12senzanmol.fit(Xtr, Ytr)
    
y_cv12senzanmol = cross_val_predict(pls_12senzanmol, Xtr, Ytr, cv=10)
    
score12cv = r2_score(Ytr, y_cv12senzanmol)
mse12cv = mean_squared_error(Ytr, y_cv12senzanmol)    

print("comp: %4d MSE: %10.8f score: %10.8f"%(ncompgraf_senzanmol, mse12cv, score12cv))
```

```python
Y_pred21senzanmol = pls_21senzanmol.predict(Xte)
Y_pred12senzanmol = pls_12senzanmol.predict(Xte)

mse_pred21 = mean_squared_error(Yte, Y_pred21senzanmol )
mse_pred12 = mean_squared_error(Yte, Y_pred12senzanmol )
RMSE_pred21 = mean_squared_error(Yte, Y_pred21senzanmol, squared = False )
RMSE_pred12 = mean_squared_error(Yte, Y_pred12senzanmol, squared = False )

diff12 = []
diff21 = []

for idx, val in enumerate(Yte):
    diff12.append(val-Y_pred12senzanmol[idx])
    diff21.append(val-Y_pred21senzanmol[idx])
    print("%10.7f %10.7f %10.7f %10.7f %10.7f"%(\
                                val, Y_pred21senzanmol[idx], diff21[-1], \
                            Y_pred12senzanmol[idx], diff12[-1]))
    
print("MSE 21 comp: ", mse_pred21) 
print("MSE 12 comp: ", mse_pred12)
score21 = r2_score(Yte, Y_pred21senzanmol )
print(" score modello a 21 componenti :", score21 )
score12 = r2_score(Yte, Y_pred12senzanmol )
print(" score modello a 12 componenti :", score12 )
print(" RMSE modello a 21 componenti :", RMSE_pred21)
print(" RMSE modello a 12 componenti :", RMSE_pred12)


plt.xlabel("True Y value")
plt.ylabel("Predicted Y values without nmol ( model built using 21 components)")

plt.scatter(Yte, Y_pred21senzanmol)

plt.show()

plt.xlabel(" True Y value")
plt.ylabel(" Predicted Y values without nmol(model built using 12 components) ")

plt.scatter(Yte, Y_pred12senzanmol)

plt.show()
```
