```python
import pandas as pd 
file_base = './data/MOE_GRID_1078_SCALED_CONCAT_dataframe.csv'
file_base_read = pd.read_csv(file_base)
desc_MOE_GRID_scaled = pd.DataFrame(file_base_read)
desc_MOE_GRID_scaled
```

```python
import numpy as np


list_of_desc = ["a_acc", \
                  "a_base", \
                  "bpol", \
                  "b_double", \
                  "b_rotN", \
                  "b_rotR", \
                  "chi1_C", \
                  "diameter", \
                  "logP(o/w)", \
                  "logS", \
                  "PEOE_VSA+0", \
                  "PEOE_VSA+3", \
                  "PEOE_VSA-0", \
                  "PEOE_VSA-2", \
                  "PEOE_VSA-5", \
                  "PEOE_VSA_POS", \
                  "SlogP_VSA0", \
                  "SlogP_VSA1", \
                  "SlogP_VSA2", \
                  "SlogP_VSA3", \
                  "SlogP_VSA4", \
                  "SlogP_VSA8", \
                  "SMR_VSA1", \
                  "SMR_VSA6", \
                  "TPSA", \
                  "vsa_acc", \
                  "vsa_other", \
                  "vsa_pol", \
                  "weinerPol"]

basicname = ["M"]

X = {}
Y = {}
setid = {}
uniqueid = {}

for bn in basicname:
    X[bn] = []
    Y[bn] = []
    setid[bn] = []
    uniqueid[bn] = []

for idx, ss in enumerate(desc_MOE_GRID_scaled["UNIQUEID"]):
    
    fdesc = desc_MOE_GRID_scaled[desc_MOE_GRID_scaled["UNIQUEID"] == ss]
    
    xval = {}
    for bn in basicname:
        xval[bn] = []

    for d in list_of_desc:
        xval["M"].append(fdesc[d].values[0])
            
    y = desc_MOE_GRID_scaled[desc_MOE_GRID_scaled["UNIQUEID"] == ss]['LOGD'].values[0]
    sid = desc_MOE_GRID_scaled[desc_MOE_GRID_scaled["UNIQUEID"] == ss]['SETID'].values[0]
    uid = desc_MOE_GRID_scaled[desc_MOE_GRID_scaled["UNIQUEID"] == ss]['UNIQUEID'].values[0]
    
    for bn in ["M"]:
        X[bn].append(xval[bn])
        Y[bn].append(y)
        setid[bn].append(sid)
        uniqueid[bn].append(uid)

for bn in basicname:
    print(bn, " : ")
    print("  ", np.asarray(Y[bn]).shape, np.asarray(X[bn]).shape ) 
```
