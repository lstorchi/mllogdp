```python
import pandas as pd 

filename = "./data/1130_alldesc.csv"

descdata = pd.read_csv(filename)

filename = "./data/data.xlsx"

ydata = pd.read_excel(filename)

if len(ydata) != len(descdata):
    print("Error")

print(ydata.head())
```

```python

list_of_desc = ["a_acc", \
                  "a_base", \
                  "bpol", \
                  "b_double", \
                  "b_rotN", \
                  "b_rotR", \
                  "chi1_C", \
                  "diameter", \
                  "logP(o/w)", \
                  "logS", \
                  "nmol", \
                  "PEOE_VSA+0", \
                  "PEOE_VSA+3", \
                  "PEOE_VSA-0", \
                  "PEOE_VSA-2", \
                  "PEOE_VSA-5", \
                  "PEOE_VSA_POS", \
                  "SlogP_VSA0", \
                  "SlogP_VSA1", \
                  "SlogP_VSA2", \
                  "SlogP_VSA3", \
                  "SlogP_VSA4", \
                  "SlogP_VSA8", \
                  "SMR_VSA1", \
                  "SMR_VSA6", \
                  "TPSA", \
                  "vsa_acc", \
                  "vsa_other", \
                  "vsa_pol", \
                  "weinerPol"]

X = []
Y = []
setid = []
for idx, ss in enumerate(ydata["NO"]):
    
    fdesc = descdata[descdata["Unique_ID"] == ss+".mol"]
    
    xval = []
    for d in list_of_desc:
        xval.append(fdesc[d].values[0])

    X.append(xval)
    Y.append(ydata[ydata["NO"] == ss]["LogD"].values[0])
    setid.append(ydata[ydata["NO"] == ss]["DatasetID"].values[0])
    print(len(Y))
    
for idx, y in enumerate(Y):
    print(setid[idx], X[idx], y)
```

```python
Xtr = []
Xte = []
Ytr = []
Yte = []
for idx, y in enumerate(Y):
    if setid[idx] == "Tr":
        Xtr.append(X[idx])
        Ytr.append(Y[idx])
    elif setid[idx] == "Te":
        Xte.append(X[idx])
        Yte.append(Y[idx])
        
print("    Test set: ", len(Yte), len(Xte))
print("Training set: ", len(Ytr), len(Xtr))
```

# Just splitted in training  and test set 

```python
# valutazione del numero di componenti principali ideale da usare per costruire il modello PLS
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_predict
import numpy as np
import matplotlib.pyplot as plt

trscores = []
trmses = []
trncomps = []
for trncomp in range(1,20):
    pls = PLSRegression(trncomp)
    pls.fit(Xtr, Ytr)
    
    y_cv = cross_val_predict(pls, Xtr, Ytr, cv=10)
    
    trscore = r2_score(Ytr, y_cv)
    trmse = mean_squared_error(Ytr, y_cv)
    
    trscores.append(trscore)
    trmses.append(trmse)
    trncomps.append(trncomp)
    print("%4d %10.8f %10.8f"%(trncomp, trmse, trscore))
    
msetrmin = np.argmin(trmses)
print("Suggested number of components tr: ", msetrmin+1)
    
plt.plot(trncomps, trscores, '-o', color='black')
plt.xlabel('number of components tr')
plt.ylabel('scores tr')
plt.xticks(trncomps)
plt.show()

plt.plot(trncomps, trmses, '-o', color='black')
plt.xlabel('number of components tr')
plt.ylabel('MSE tr')
plt.xticks(trncomps)
plt.show()
```

```python
# il numero di componenti suggerito ricerca che ho fatto è di 16 componenti principali, anche se probabilmente sarebbe possibileusarne di meno volendo( penso che 6-7 componenti potrebbero comunque andare bene )
```

```python
ncompsugg = 16

pls_16 = PLSRegression(ncompsugg)
pls_16.fit(Xtr, Ytr)
    
y_cv16 = cross_val_predict(pls_16, Xtr, Ytr, cv=10)
    
score16cv = r2_score(Ytr, y_cv16)
mse16cv = mean_squared_error(Ytr, y_cv16)    

print("comp: %4d MSE: %10.8f score: %10.8f"%(ncompsugg, mse16cv, score16cv))
```

```python
ncompgraf = 7

pls_7 = PLSRegression(ncompgraf)
pls_7.fit(Xtr, Ytr)
    
y_cv7 = cross_val_predict(pls_7, Xtr, Ytr, cv=10)
    
score7cv = r2_score(Ytr, y_cv7)
mse7cv = mean_squared_error(Ytr, y_cv7)    

print("comp: %4d MSE: %10.8f score: %10.8f"%(ncompgraf, mse7cv, score7cv))
```

```python
Y_pred16 = pls_16.predict(Xte)
Y_pred7 = pls_7.predict(Xte)

mse_pred16 = mean_squared_error(Yte, Y_pred16 )
mse_pred7 = mean_squared_error(Yte, Y_pred7 )
RMSE_pred16 = mean_squared_error(Yte, Y_pred16, squared = False )
RMSE_pred7 = mean_squared_error(Yte, Y_pred7, squared = False )

diff7 = []
diff16 = []

for idx, val in enumerate(Yte):
    diff7.append(val-Y_pred7[idx])
    diff16.append(val-Y_pred16[idx])
    print("%10.7f %10.7f %10.7f %10.7f %10.7f"%(\
                                val, Y_pred16[idx], diff16[-1], \
                            Y_pred7[idx], diff7[-1]))
    
print("MSE 16 comp: ", mse_pred16) 
print("MSE 7 comp: ", mse_pred7)
score16 = r2_score(Yte, Y_pred16 )
print(" score modello a 16 componenti :", score16 )
score7 = r2_score(Yte, Y_pred7 )
print(" score modello a 7 componenti :", score7 )
print(" RMSE modello a 16 componenti :", RMSE_pred16)
print(" RMSE modello a 7 componenti :", RMSE_pred7)


plt.xlabel("True Y value")
plt.ylabel("Predicted Y values ( model built using 16 components)")

plt.scatter(Yte, Y_pred16)

plt.show()

plt.xlabel(" True Y value")
plt.ylabel(" Predicted Y values (model built using 7 components) ")

plt.scatter(Yte, Y_pred7)

plt.show()
```
