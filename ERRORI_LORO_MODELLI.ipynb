```python
import pandas as pd
file = './data/RISULTATI_LORO_MODELLI_togliriaga1.csv'
res_theirPLS = pd.read_csv(file)
res_theirPLS_data = pd.DataFrame(res_theirPLS)
res_theirPLS_data
```

```python
res_theirPLS_data.columns
```

```python
file2 = './data/MOE_GRID_1078.xlsx'
read_desc = pd.read_excel(file2)
desc_1078 = pd.DataFrame(read_desc)
desc_1078
```

```python
merged_dataframe_MOEnostri = pd.merge(res_theirPLS_data, desc_1078, on='UNIQUEID')
merged_dataframe_MOEnostri
```

```python
True_value = []
PLS_LORO = []
SETID = []
SVM = []
DS =[]
Marvin_Weight = []
M_VG = []
M_KLOP = []
M_PHYS = []
UNIQUEID = []
for idx, unique in enumerate(merged_dataframe_MOEnostri["UNIQUEID"]):
    True_value.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['True_val'].values[0])
    PLS_LORO.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['PLS'].values[0])
    SETID.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['Setids'].values[0])
    SVM.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['SVM'].values[0])
    DS.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['DS'].values[0])
    Marvin_Weight.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['Marvin_Weight'].values[0])
    M_VG.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['M_VG'].values[0])
    M_KLOP.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['M_KLOP'].values[0])
    M_PHYS.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['M_PHYS'].values[0])
    UNIQUEID.append(merged_dataframe_MOEnostri[merged_dataframe_MOEnostri["UNIQUEID"] == unique]['UNIQUEID'].values[0])

print(len(True_value))
print(len(PLS_LORO))
print(len(SETID))
print(len(SVM))
print(len(DS))
print(len(Marvin_Weight))
print(len(M_VG))
print(len(M_KLOP))
print(len(M_PHYS))
```

```python
true_tr = []
true_te = []
PLS_tr = []
PLS_te = []
SETID_Tr = []
SETID_Te = []

for idx,trval in enumerate (True_value):
    if SETID[idx]=='Tr':
        true_tr.append(True_value[idx])
        PLS_tr.append(PLS_LORO[idx])
        SETID_Tr.append(SETID[idx])
        
    if SETID[idx]=='Te':
        true_te.append(True_value[idx])
        PLS_te.append(PLS_LORO[idx])
        SETID_Te.append(SETID[idx])
        
print(len(true_te))
print(len(true_tr))
print(len(PLS_te))
print(len(PLS_tr))
```

```python
diff_PLS_tr = []


for idx, val in enumerate(true_tr):
    diff_PLS_tr.append(val-PLS_tr[idx])
    print("%10.7f %10.7f %10.7f"%(\
                                val, PLS_tr[idx], diff_PLS_tr[-1] )) 
```

```python
diff_PLS_te = []


for idx, val in enumerate(true_te):
    diff_PLS_te.append(val-PLS_te[idx])
    print("%10.7f %10.7f %10.7f"%(\
                                val, PLS_te[idx], diff_PLS_te[-1] ))
```

```python
diff_PLS_tot = []


for idx, val in enumerate(True_value):
    diff_PLS_tot.append(val-PLS_LORO[idx])
    print("%10.7f %10.7f %10.7f"%(\
                                val, PLS_LORO[idx], diff_PLS_tot[-1] ))
```

```python
print(len(diff_PLS_te))
print(len(diff_PLS_tr))
print(len(diff_PLS_tot))
```

```python
import numpy as np
stdev_diff_PLS_tr = np.std(diff_PLS_tr)
stdev_diff_PLS_te = np.std(diff_PLS_te)
stdev_diff_PLS_tot = np.std(diff_PLS_tot)
print("%10.7f %10.7f %10.7f"%(stdev_diff_PLS_te, stdev_diff_PLS_tr, stdev_diff_PLS_tot ))
abs_diff_PLS_tr = np.absolute(diff_PLS_tr)
avg_abs_PLS_tr = np.average(abs_diff_PLS_tr)
abs_diff_PLS_te = np.absolute(diff_PLS_te)
avg_abs_PLS_te = np.average(abs_diff_PLS_te)
abs_diff_PLS_tot = np.absolute(diff_PLS_tot)
avg_abs_PLS_tot = np.average(abs_diff_PLS_tot)
print("%10.7f %10.7f %10.7f"%(avg_abs_PLS_te, avg_abs_PLS_tr, avg_abs_PLS_tot ))
stdev_abs_PLS_tr = np.std(abs_diff_PLS_tr)
stdev_abs_PLS_te = np.std(abs_diff_PLS_te)
stdev_abs_PLS_tot = np.std(abs_diff_PLS_tot)
print("%10.7f %10.7f %10.7f"%(stdev_abs_PLS_te, stdev_abs_PLS_tr, stdev_abs_PLS_tot ))

```
